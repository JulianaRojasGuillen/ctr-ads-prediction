# Exploring DataSet

```python
#To display all columns
df.columns
```

```python
# Get the number of unique values in each column
unique_counts = df.nunique()
print(unique_counts)
```

```python
# Get the number of missing values in each column
missing_counts = df.isnull().sum()
print(missing_counts)
```

```python
# Get the data type of each column
data_types = df.dtypes
print(data_types)
```

```python
# Examine imbalance in the target variable 'click'
click_values = df['click'].value_counts()
print(click_values)

print("Imbalance ratio: {}".format(float(len(df[df['click']==0]))/len(df[df['click']==1])))
```

```python
# Calculating the click through rate in our sample
CTR = df['click'].mean()
print(CTR)
```

# Preprocessing

```python
# Deleting id column
df = df.drop('id', axis=1)
df.head()
```

```python
from datetime import datetime

df['hour'] = df['hour'].map(lambda x: datetime.strptime(str(x), "%y%m%d%H"))
df['day_of_week'] = df['hour'].map(lambda x: x.weekday())
df['hour'] = df['hour'].map(lambda x: x.hour)

df.head()
```
